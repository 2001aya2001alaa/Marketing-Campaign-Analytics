# Feature Engineering For a Predictive ML Model

```python
# Classify The Customer Based on there ages (Young, Middle, Old)
temp = []
for row in data["Age"]:
    if row <= 30 : temp.append("Young")
    elif (row > 30 and row <= 55) : temp.append("Middle")
    elif row > 55 : temp.append("Old")    
```

```python
data["Age_classify"] = temp
data
```
