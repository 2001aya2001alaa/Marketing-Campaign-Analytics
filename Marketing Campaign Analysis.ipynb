```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Load The Data

```python
df = pd.read_csv('Downloads/marketing_campaign.csv')
```

```python
df
```

# Clean The Data

```python
df.duplicated().sum()
```

```python
#Remove Duplicates
df = df.drop_duplicates(subset=["Year_Birth", "Education", "Marital_Status", "Income", "Dt_Customer", "Recency"], keep = 'first')
```

```python
df
# 201 Duplicates
```

```python
# Delete Columns with only one value
df = df.drop(['Z_CostContact', 'Z_Revenue'], axis=1)
```

```python
df
```

```python
# Delete Columns That will not be used
df = df.drop(['Complain'], axis=1)
```

```python
# Handle Null Values
df.isnull().sum()
```

```python
df = df.dropna()
```

# Prepare The Data 

```python
# Calculate the age of customer in 2014 To cluster the customer based on their ages 

df["Age"] = 2014 - df["Year_Birth"]
df
```

```python
#Create Month and Year For Analysis

df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])

df['Year'] = pd.DatetimeIndex(df['Dt_Customer']).year
df['Month'] = pd.DatetimeIndex(df['Dt_Customer']).month
df['Quarter'] = pd.DatetimeIndex(df['Dt_Customer']).quarter

df
```

```python
data = df
```

```python
data = data.drop(columns=['Year_Birth', 'Dt_Customer'])
```

```python
data.rename(columns = {'MntWines':'Mnt_Wine', 'MntFruits':'Mnt_Fruit', 'MntMeatProducts':'Mnt_Meat',
                       'MntFishProducts':'Mnt_Fish', 'MntSweetProducts':'Mnt_Sweet', 'MntGoldProds':'Mnt_Gold'}, 
            inplace = True)
```
